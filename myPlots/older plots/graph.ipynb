```python
# importing package
import matplotlib.pyplot as plt
import numpy as np
```

The next set of graphs are stacked bars for:
* DDR4-DDR4  With different link latencies
* RO-Miss-Clean
* RO-Miss-Dirty
* WO-Miss-Dirty

```python
# create data
x = ['~1 (ns)', '200 (ns)', '1 (us)', '5 (us)', '10 (us)']
y1 = np.array([1175.7, 1514.5, 957.1, 121.5, 169.8]) # loc read
y2 = np.array([73.5, 274.9, 1075.9, 5081.3, 10077.0]) # far read
y3 = np.array([988.6, 1305.2, 1203.9, 90.1, 105.9]) # loc write
  
# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')

plt.xlabel("Link Latency")
plt.ylabel("Time (ns)")
plt.legend(["AvgLocRdTime", "AvgFarRdTime", "AvgLocWrTime"])
plt.title("RO, Miss-Clean")
plt.show()
```

```python
# ART = average response time
input_inj_rate = [60, 30, 15, 7.5, 3.75, 1.875] # GB/s
ART_RO = [2923, 2923, 2914, 2816, 2723, 2667] 
ART_R67 = [1961, 1961, 1950, 1878, 1831, 1788]
ART_WO = [18, 18, 14, 10, 10, 10]
  
# plot lines
plt.plot(input_inj_rate, ART_RO, label = "RO")
plt.scatter(input_inj_rate, ART_RO)

plt.plot(input_inj_rate, ART_R67, label = "R67")
plt.scatter(input_inj_rate, ART_R67)

plt.plot(input_inj_rate, ART_WO, label = "WO")
plt.scatter(input_inj_rate, ART_WO)

plt.title("DDR4-DDR4 / Miss-Clean / 2.5 micro-sec Link")
plt.xlabel("Injected Input BW (GB/s)")
plt.ylabel("Average Response Time (ns)")
plt.legend()
plt.show()
```
